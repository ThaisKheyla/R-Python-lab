#Desafio
Encontrar as ações que são recomendadas por todas as corretoras.

```python
agoraSet ={
"empresas": {"Itaúsa","Ecorodovias","Taesa","B3","Vale"},
"tickets": {"ITSA4","ECOR3","TAEE11","B3SA3","VALE3"}}

ativaSet = {
"empresas": {"B3", "Bradesco", "BB Seguridade", "BR Distribuidora", "Taesa", "CTEEP", "Vale", "Telefônica Brasil"},
"tickets": {"B3SA3", "BBDC4", "BBSE3", "BRDT3", "TAEE11", "TRPL4", "VALE3", "VIVT3"}}

genialSet = {
"empresas": {"CPFL", "Minerva", "Cyrela", "Randon", "CTEEP"},
"tickets": {"CPFE3", "BEEF3", "CYRE3", "SAPT4", "TRPL4"}}

easynvestSet = {
"empresas": {"B3", "Brasil Agro", "Coca-Cola", "Taesa", "Vale", "Copel", "Itaúsa", "Ambev"},
"tickets": {"B3SA3", "AGRO3", "COCA34", "TAEE11", "VALE3", "CPLE11", "ITSA4", "ABEV3"}
}

eliteSet = {
"empresas": {"Bradesco", "BB Seguridade", "Banrisul", "Engie", "Itaúsa", "Sanepar", "Taesa", "CTEEP", "Telefônica Brasil", "Vale"},
"tickets": {"BBDC4", "BBSE3", "BRSR6", "EGIE3", "ITSA4", "SAPR11", "TAEE11", "TRPL4", "VIVT3", "VALE3"}
}

guideSet = {
"empresas": {"Alupar", "Banco do Brasil", "Cyrela", "CPFL", "Klabin", "Porto Seguro", "Tim", "Vale"},
"tickets": {"ALUP11", "BBAS3", "CYRE3", "CPFE3", "KLBN11", "PSSA3", "TIMS3", "VALE3"}
}

novaFuturaSet = {
"empresas": {"B3", "Cyrela", "Gerdau", "Vivo", "CTEEP"},
"tickets": {"B3SA3", "CYRE3", "GGBR4", "VIVT3", "TRPL4"}
}


oramaSet = {
"empresas": {"Banco ABC", "Bradesco", "Minerva", "CESP", "Engie"},
"tickets": {"ABCB4", "BBDC4", "BEEF3", "CESP6", "EGIE3"}
}

acoes_comuns = agoraSet["tickets"].intersection(
    ativaSet["tickets"],
    genialSet["tickets"],
    easynvestSet["tickets"],
    eliteSet["tickets"],
    guideSet["tickets"],
    novaFuturaSet["tickets"],
    oramaSet["tickets"]
)


print("Ações recomendadas por todas as corretoras:", acoes_comuns)
```

## Identificar ações exclusivas de cada corretora

Encontrar as ações que são recomendadas por apenas uma corretora específica.


```python
todosTickets = (
    agoraSet["tickets"]
    | ativaSet["tickets"]
    | genialSet["tickets"]
    | easynvestSet["tickets"]
    | eliteSet["tickets"]
    | guideSet["tickets"]
    | novaFuturaSet["tickets"]
    | oramaSet["tickets"]
)

agora_exclusivas = agoraSet["tickets"] - (todosTickets - agoraSet["tickets"])
ativa_exclusivas = ativaSet["tickets"] - (todosTickets - ativaSet["tickets"])
genial_exclusivas = genialSet["tickets"] - (todosTickets - genialSet["tickets"])
easynvest_exclusivas = easynvestSet["tickets"] - (todosTickets - easynvestSet["tickets"])
elite_exclusivas = eliteSet["tickets"] - (todosTickets - eliteSet["tickets"])
guide_exclusivas = guideSet["tickets"] - (todosTickets - guideSet["tickets"])
novaFutura_exclusivas = novaFuturaSet["tickets"] - (todosTickets - novaFuturaSet["tickets"])
orama_exclusivas = oramaSet["tickets"] - (todosTickets - oramaSet["tickets"])

print(f"Ações exclusivas da Agora: {agora_exclusivas}")
print(f"Ações exclusivas da Ativa: {ativa_exclusivas}")
print(f"Ações exclusivas da Genial: {genial_exclusivas}")
print(f"Ações exclusivas da Easynvest: {easynvest_exclusivas}")
print(f"Ações exclusivas da Elite: {elite_exclusivas}")
print(f"Ações exclusivas da Guide: {guide_exclusivas}")
print(f"Ações exclusivas da Nova Futura: {novaFutura_exclusivas}")
print(f"Ações exclusivas da Orama: {orama_exclusivas}")
print("eu odeio minha vidaaaaaa")
```

Determinar se todas as ações recomendadas por uma corretora estão incluídas nas recomendações de outra corretora


```python
corretoras = {
    "Agora": agoraSet,
    "Ativa": ativaSet,
    "Genial": genialSet,
    "Easynvest": easynvestSet,
    "Elite": eliteSet,
    "Guide": guideSet,
    "Nova Futura": novaFuturaSet,
    "Orama": oramaSet
}

for nome1, dados1 in corretoras.items():
    for nome2, dados2 in corretoras.items():
        eh_subconjunto = dados1["tickets"].issubset(dados2["tickets"])
        print(f"A carteira da {nome1} é subconjunto da carteira da {nome2}: {eh_subconjunto}")

```
