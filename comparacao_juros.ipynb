# Comparação de Juros Simples e Compostos

*Este script em R realiza uma simulação financeira com o objetivo de comparar o desempenho do Regime de Juros Simples e do Regime de Juros Compostos para um dado investimento. O programa calcula o retorno final, determina a modalidade mais vantajosa e gera uma visualização gráfica da evolução dos montantes ao longo do tempo*

# Definição das Variáveis de Entrada

*Esta etapa inicial é dedicada à coleta dos parâmetros essenciais fornecidos pelo usuário, garantindo a flexibilidade da análise:*

- **Capital Inicial (C):** O valor principal (Principal) a ser investido ou capitalizado.

- **Taxa de Juros (i):** A taxa de juros aplicada por período (mensal), inserida em formato percentual.

- **Tempo (t):** A duração total do investimento, em meses.

*Tratamento de Dados: É crucial que a taxa de juros (i) seja imediatamente convertida para a sua forma decimal (i / 100) para que possa ser utilizada corretamente nos modelos matemáticos de capitalização*

```r
# Entrada de dados do usuário
cat("=== Comparação de Juros Simples e Compostos ===\n")

P <- as.numeric(readline("Digite o valor inicial investido (R$): "))
i <- as.numeric(readline("Digite a taxa de juros mensal (%): "))
t <- as.integer(readline("Digite o tempo do investimento (em meses): "))

# Converter taxa percentual para decimal
i <- i / 100
```

# Cálculo de Montante e Lucro

*As fórmulas de capitalização são aplicadas para determinar o retorno de cada regime ao final do período $t$:*

image.png

```r
# Cálculos
M_simples <- P * (1 + i * t)

M_compostos <- P * (1 + i)^t

# Lucros
lucro_simples <- M_simples - P
lucro_compostos <- M_compostos - P
```

# Preparação do Vetor de Progressão Temporal

*Para a visualização gráfica, é necessário calcular o valor acumulado em cada intervalo de tempo (mês a mês, de $0$ a $t$).*

- O vetor meses define o eixo $X$ (o tempo).
- Os vetores valores_simples e valores_compostos definem o eixo $Y$ (o montante), ilustrando a trajetória linear e exponencial, respectivamente, usando o Capital $P$ como base.

```r
# Geração dos valores mês a mês
meses <- 0:t
valores_simples <- P * (1 + i * meses)
valores_compostos <- P * (1 + i)^meses
```

# Visualização Gráfica da Evolução

*O gráfico de linhas é fundamental para demonstrar o diferencial dos regimes, especialmente o efeito da capitalização composta em longo prazo.*

- A curva Azul representa a progressão constante dos Juros Simples.

- A curva Vermelha demonstra a aceleração característica dos Juros Compostos, onde a diferença de performance se torna mais evidente com o aumento do tempo.

```r
# Gráfico
plot(meses, valores_simples, type = "l", col = "blue", lwd = 2,
     xlab = "Meses", ylab = "Valor acumulado (R$)",
     main = "Comparação: Juros Simples vs Juros Compostos")
lines(meses, valores_compostos, col = "red", lwd = 2)
legend("topleft", legend = c("Juros Simples", "Juros Compostos"),
       col = c("blue", "red"), lwd = 2)
```

# Apresentação dos Resultados e Conclusão

*A seção final exibe o resumo dos montantes e lucros calculados e fornece o diagnóstico da simulação.*

O uso de estruturas condicionais (if/else) permite identificar objetivamente o investimento mais rentável e quantificar a diferença exata de lucro entre as duas modalidades, finalizando a análise comparativa:

```r
# Resultado final
cat("\n=== RESULTADOS FINAIS ===\n")
cat(sprintf("Montante (Juros Simples): R$ %.2f\n", M_simples))
cat(sprintf("Lucro (Juros Simples): R$ %.2f\n", lucro_simples))
cat(sprintf("Montante (Juros Compostos): R$ %.2f\n", M_compostos))
cat(sprintf("Lucro (Juros Compostos): R$ %.2f\n", lucro_compostos))

# Comparação e diferença
if (M_compostos > M_simples) {
  cat("\nO investimento com JUROS COMPOSTOS foi mais vantajoso!\n")
  cat(sprintf("Diferença de lucro: R$ %.2f\n", lucro_compostos - lucro_simples))
} else if (M_simples > M_compostos) {
  cat("\nO investimento com JUROS SIMPLES foi mais vantajoso!\n")
  cat(sprintf("Diferença de lucro: R$ %.2f\n", lucro_simples - lucro_compostos))
} else {
  cat("\nAmbos os investimentos tiveram o mesmo rendimento.\n")
}
```
